# Mean of 10 trials

```python
li_state1 = []
li_state1.append(x1[len(x1)-2500:].values)
li_state1.append(x2[len(x2)-2500:].values)
li_state1.append(x3_1[len(x3_1)-2500:].values)
li_state1.append(x4[len(x4)-2500:].values)
li_state1.append(x5[len(x5)-2500:].values)
li_state1.append(x6[len(x6)-2500:].values)
li_state1.append(x7_1[len(x7_1)-2500:].values)
li_state1.append(x8[len(x8)-2500:].values)
li_state1.append(x9[len(x9)-2500:].values)
li_state1.append(x10[len(x10)-2500:].values)
ncols = len(li_state1)
nrows = len(li_state1[0])
results_x = nrows*[0] # sums per column, afterwards avgs
num_elem = ncols
for row in range(nrows):
    for col in range(ncols):
        results_x[row] += li_state1[col][row]
    results_x[row] /= num_elem


li_state2 = []
li_state2.append(force1[len(force1)-2500:].values)
li_state2.append(force2[len(force2)-2500:].values)
li_state2.append(force3_1[len(force3_1)-2500:].values)
li_state2.append(force4[len(force4)-2500:].values)
li_state2.append(force5[len(force5)-2500:].values)
li_state2.append(force6[len(force6)-2500:].values)
li_state2.append(force7_1[len(force7_1)-2500:].values)
li_state2.append(force8[len(force8)-2500:].values)
li_state2.append(force9[len(force9)-2500:].values)
li_state2.append(force10[len(force10)-2500:].values)
ncols = len(li_state2)
nrows = len(li_state2[0])
results_force = nrows*[0] # sums per column, afterwards avgs
num_elem = ncols
for row in range(nrows):
    for col in range(ncols):
        results_force[row] += li_state2[col][row]
    results_force[row] /= num_elem

```
